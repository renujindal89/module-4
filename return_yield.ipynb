```python
return

When you use return to give back a list (or any collection),
the entire list is created and stored in memory at once.

This can consume a lot of memory if the list is large.

➤ yield

When you use yield, the function becomes a generator.

It produces one value at a time, only when needed (lazy evaluation).

It doesn’t keep the entire sequence in memory — just the current state.
```

```python
def counter():
    print("hello")
    print("second")
    return 
    print("hru")
counter()    
    
```

```python
def counter():
    for i in range(5):
        print(i)
    return
    print("after loop")
counter()        
```

```python
def counter():
    print("hello")
    yield 0 
    print("hru")
    yield 1
    print("welcomr")
    yield 2
counter()   
```

```python
gen=counter()
next(gen)  # runs untill first yield 
```

```python
next(gen)
```
