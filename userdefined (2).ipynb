```python
#sum of numbers
def add_all(*args):  #1,2,3
    total = 0
    for num in args: #1,2,3
        total += num
    return total

print(add_all(1, 2, 3))        # 6
print(add_all(5, 10, 15, 20))  # 50


#args is a tuple like (1, 2, 3)
```

```python
def calculate_sum(start_value,last_value,*numbers): #positional+*args
    print(start_value)
    print(last_value)
    total = start_value
    for num in numbers:
        total += num
    return total

# Pass multiple additional arguments.
print(calculate_sum( 1,1,1, 2, 3))
# Output: 16

# Pass no additional arguments.
print(calculate_sum(10,20))
# Output: 10
```

# enclosing scope

```python
In programmingâ€”especially in Pythonâ€”an enclosing scope variable refers to a variable that:

Is defined in an outer (enclosing) function, and

Is used inside an inner (nested) function.
```

```python
def outer():
    x = 10   # <-- enclosing scope variable

    def inner():
        return x  # inner function uses variable from the enclosing scope

    return inner

f = outer()
print(f())   # prints 10
```

```python
def fun():
    print("inside function")
fun() 
def fun1(n):
    return(n)
print(2+fun1(4))    
```

```python
explanation
When outer() runs:

It creates a local variable x = 10.

It defines an inner function inner() that returns x.

outer() then returns the function inner (not executed, just returned as an object).
When f() executes:

It runs the inner function.

inner returns the value of x it remembered from outer.

That value is 10, so print outputs:
```

```python
def login(user, password):
    def check_credentials():
        return user == "admin" and password == "1234"

    if check_credentials():   # we call the inner function here
        return "Access granted"
    else:
        return "Access denied"

print(login("admin", "1234"))
```

# Global scope

```python
A variable defined at the top level of a script or module has a global scope
and is accessible throughout the program
```

```python
z = 40  # `z` has a global scope

def my_function():
    print(z)  # Can access `z` from the global scopeprint(z)
my_function()
print(z)
```

```python
# wants to modify the global and enclosing variable
```

```python
z = 40  # global

def my_function():
    z += 1    # tries to modify z
    print(z)

my_function()
```

```python
When you write z += 1 inside a function, Python automatically treats z as a local variable.
But since you did not define a local z before using +=, Python complains.
How to fix it (if you want to modify the global variable)

Use the global keyword:
```

```python
z = 40  # `z` has a global scope

def my_function():
    global z  # tell Python to use the global z
    z += 1  # tries to modify z
    print(z)
my_function()
print(z)
```

```python
Using nonlocal (for enclosing function variables)

nonlocal lets you modify a variable from the enclosing (non-global) scope â€”
meaning a variable defined in an outer function but not global.
```

```python
def outer():
    x = 10   # enclosing scope variable

    def inner():
        nonlocal x   # use x from outer()
        x += 1
        print("inner:", x)

    inner()
    print("outer:", x)

outer()
```
